```python
nums = [100,6,200,1,8,12]

longest = 0

for num in nums:

    current = num
    length = 1

    while current + 1 in nums:

        current += 1
        print(current)
        length += 1
        # print(length)
        # print(longest)

    longest = max(longest, length)

print(longest)
```

```python
def swap_case(s):
    str=''
    for char in s:
        if char==char.lower():
            str+=char.upper()
        else:
            str += char.lower()    
        
    return str

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```
