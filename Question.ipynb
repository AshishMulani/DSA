```python
# Create the chunk grammer for regular expression
grammer = "NP:{<DT><JJ><NN>}"
```

```python
sent = 'It was a beautiful game by him. An amazing play was witnessed.'
```

```python
parser = RegexpParser(grammer)
```

```python
tree = parser.parse(pos_tag(word_tokenize(sent)))
```

```python
tree
```

```python
tree.draw()
```

```python
grammer = r"NP:{<DT>?<JJ>*<NN>+}"
# <DT>? Zero or One Determiner 
# <JJ>* Zero or more adjectives 
# <NN>+ One or nouns
```

```python
sent = 'It was a beautiful fine game by him. An amazing play was witnessed.'
parser = RegexpParser(grammer)
tree = parser.parse(pos_tag(word_tokenize(sent)))
```

```python
tree
```

```python
def clean_text(sent):
    # tokenize the data 
    tokens1 = word_tokenize(sent)
    # remove punctuation and numbers 
    tokens2 = [token for token in tokens1 if token.isalpha()]
    # remove stopwords 
    tokens3 = [token.lower() for token in tokens2 
           if token.lower() not in stopwords.words('english')]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4
```

```python
import pandas as pd
import tensorflow as tf
import torch, evaluate
import numpy as np
from datasets import Dataset
from sklearn.model_selection import train_test_split
from transformers import (
        BertTokenizer, BertForSequenceClassification,
          TrainingArguments, Trainer, pipeline)
```

```python
df = pd.read_csv('SMSSpamCollection', sep='\t',
                 names= ['label', 'text'])
```

```python
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
train, test = train_test_split(df, test_size=0.2, random_state=0)
```

### Convert Pandas df to huggingface datasets

```python
train_dataset = Dataset.from_pandas(train)
test_dataset = Dataset.from_pandas(test)
```

```python
type(train_dataset)
```

### Tokenization

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def tokenizer_function(examples):
      return tokenizer(examples['text'],
                  padding='max_length',truncation=True,
                  max_length=128)
```

```python
train_dataset = train_dataset.map(tokenizer_function, batched=True)
test_dataset = test_dataset.map(tokenizer_function, batched=True)
```

```python
train_dataset = train_dataset.remove_columns(['text', '__index_level_0__'])
test_dataset = test_dataset.remove_columns(['text','__index_level_0__'])
```

```python
model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased', num_labels=2)
```

```python
metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    eval_strategy='epoch',
    save_strategy='epoch'

)   ## Auto Generated first 4 atleast imp
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)
trainer.train()
```

```python
# save model
model.save_pretrained('./model')
tokenizer.save_pretrained('./tokenizer')
```

#### CHATBOT

```python
import nltk
import random
```

```python
responses = {
    'greeting': ['Hello!', 'Hi There!', 'Hey!', 'Howdy!'],
    'goodbye' : ['Good bye!', 'See you later!', 'Take care.', 'Bye!'],
    'thanks'  : ['You are welcome.', 'Glad to help', 'Anytime', 'Happy to assist'],
    'default' : ['Sorry I didnt understand', 'Can you tell again?',
                 'I am not sure.']
}
```

```python
patterns = {
    'hi': 'greeting',
    'hello': 'greeting',
    'hey': 'greeting',
    'bye': 'goodbye',
    'thanks': 'thanks',
    'thank you': 'thanks',
}
```

```python
def respond(user_input):
    for word in nltk.tokenize.word_tokenize(user_input.lower()):
        if word in patterns:
            intent = patterns[word]
            return random.choice(responses[intent])
    return random.choice(responses['default'])
```

```python
respond('hi')
respond('How are you?')
```

```python
def chat():
    print('Chatbot: How can I help you today? (Type "exit" to quit)')
    while True:
        user_input = input('You: ')
        if user_input == 'exit':
            break
        response = respond(user_input)
        print("You: ",user_input)
        print('Chatbot: ', response)

# [7]: 
chat()
```

### Flask

```python
from flask import Flask, render_template, request
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import joblib

app = Flask(__name__)

ps = PorterStemmer()
swords = stopwords.words('english')
def clean_text(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4

classifier = joblib.load('model.bin')
tfidf = joblib.load('preprocess.bin')

@app.route('/')
def student():
    return render_template('spamdetector.html')

@app.route('/spamfinder', methods=['GET', 'POST'])
def result():
    if request.method == 'POST':
        data = dict(request.form)
        message = tfidf.transform([data['message']])
        data['result'] = classifier.predict(message)[0]
        return render_template('spamoutput.html', data=data)

if __name__ == '__main__':
    app.run(debug = True)
```
